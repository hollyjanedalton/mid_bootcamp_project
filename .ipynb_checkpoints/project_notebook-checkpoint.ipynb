#### If a customer doesn't own their home ("own_your_home" = No), replace #_homes_owned with 0

```python
#in original df, where "own_your_home" = No, change "#_homes_owned" to 0
#there should at least be SOME customers who don't own their home, doesn't make sense for ALL customers
#to own at least one home. Also seems odd that only  1/3 of customers don't own a home. 

#df['#_homes_owned'] = df['#_homes_owned'].replace(['old value'],'new value')


df['#_homes_owned'] = np.where(df['own_your_home'] == 'No', 0, df['#_homes_owned'])
```
