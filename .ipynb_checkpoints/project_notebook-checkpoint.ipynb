```python
#change titles to lower case 
lower = []
for i in df.columns:
        lower.append(i.lower())
df.columns = lower

df
```

```python
#Puts underscore instead of spaces
under_ = []
for i in df.columns:
        if ' ' in i:
            under_.append(i.replace(' ', '_'))
        else:
            under_.append(i)
df.columns = under_   
df
```
