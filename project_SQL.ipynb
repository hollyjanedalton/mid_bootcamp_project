```python
import pandas as pd
import numpy as np
import pymysql

from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)
```

```python
from getpass import getpass 
```

```python
password = getpass()
```

```python
#creating the database credit_card_classification in the SQL server

connection_string = 'mysql+pymysql://root:'+password+'@localhost/credit_card_data'
engine=create_engine(connection_string)
```

```python
credit_card_classification = pd.read_csv('df_cleaned.csv', index_col=[0])
credit_card_classification.rename(columns = {'#_credit_cards_held' : 'credit_cards_held', '#_homes_owned' : 'homes_owned', '#_bank_accounts_open' : 'bank_accounts_open'}, inplace = True)
credit_card_classification
```

```python
#creating the table in the database

credit_card_classification.to_sql(name='credit_card_classification', con=engine, if_exists='replace', index=False)
```

```python
credit_card_data = credit_card_classification
credit_card_data
```

```python
credit_card_data.to_sql(name='credit_card_data', con=engine, if_exists='replace', index=False)
```
